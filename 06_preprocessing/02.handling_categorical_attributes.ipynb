```python
import pandas as pd
import numpy as np
```

# 1. LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
meslek_encoded_model = label_encoder.fit(df['meslek'])
```

```python
meslek_encoded = meslek_encoded_model.transform(df['meslek'])
```

```python
meslek_encoded
```

```python
df['meslek_encoded'] = meslek_encoded
df.sort_values('meslek_encoded').head(17)
```

```python
sehir_encoded_model = label_encoder.fit(df['sehir'])
sehir_encoded = meslek_encoded_model.transform(df['sehir'])
df['sehir_encoded'] = sehir_encoded
df.sort_values('sehir_encoded').head(17)
```

# 2. OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot = OneHotEncoder(categories='auto')
```

```python
one_hot_model = one_hot.fit(df[['meslek_encoded','sehir_encoded']])
```

```python
one_hot_model
```

```python
one_hot_all = one_hot_model.transform(df[['meslek_encoded','sehir_encoded']])
```

```python
type(one_hot_all)
```

```python
one_hot_all.toarray()
```

```python
# OneHotEncoded nitelikleri ana dataframe ile birleştirmek için dataframe'e çevirme
df_onehot = pd.DataFrame(one_hot_all.toarray())
```

```python
# Bu yöntemle birleşme indeksler üzerinden gerçekleşir.
df_join = df.join(df_onehot, how='inner')
```

```python
df_join.head(17)
```

# LabelBinarizer: LabelEncoder ve OneHotEncoder bir arada

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb_binarizer = LabelBinarizer()
```

```python
lb_bin_meslek = lb_binarizer.fit_transform(df['meslek'])
```

```python
lb_bin_meslek
```

```python
lb_bin_meslek_df = pd.DataFrame(lb_bin_meslek)
```

```python
lb_bin_meslek_df.head()
```

```python
lb_bin_sehir = lb_binarizer.fit_transform(df['sehir'])
```

```python
lb_bin_sehir 
```

```python
# Sütun isimlerinin aynı olmaması için
lb_bin_sehir_df = pd.DataFrame(lb_bin_sehir, columns=['_0','_1','_2','_3','_4','_5'])
lb_bin_sehir_df.head()
```

```python
df_join2 = df.join(lb_bin_meslek_df, how="inner")
df_join2.head()
```

```python
df_join3 = df_join2.join(lb_bin_sehir_df, how="inner")
df_join3.head()
```
